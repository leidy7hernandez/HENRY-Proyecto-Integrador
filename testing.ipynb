#### Filtrar datasets 

```python
# Eliminar stocks incompletos. Queremos acciones que coticen continuamente desde el 2013
def filter_complete_stocks(dataframe):
    df = dataframe.copy()
    #Convert date column to datetime
    df['Date'] = pd.to_datetime(df['Date'])
    #Set the Date as index
    df.set_index('Date', inplace=True)
    #Group the dataframes by ticker
    groups = df.groupby('Ticker')


    #Get the mode of number of rows to filter stocks that has less than that
    import statistics
    n_rows= groups.size()
    mode_rows = statistics.mode(n_rows)


    #Create a function to determine if the stock has more or equal rows than the mode
    def is_complete(group):
        return len(group) >= mode_rows
    

    #Get the begin date from one of the complete stocks
    beginning_date = None
    for ticker, group in groups:
        if is_complete(group):
            beginning_date = group.index.min()
            break


    #Create a function to determine if the data is continuous
    def is_continuous(group):
        return (group.index.min() == beginning_date) and (is_complete(group))


    #Apply it to filter the groups
    filtered_groups = groups.filter(is_continuous)
    filtered_df = filtered_groups.reset_index()

    return filtered_df
```

#### Calcular el retorno mensual por stock

```python
#Calcular el retorno mensual de cada stock

import pandas as pd

def calculate_monthly_returns(dataframe):
    df = dataframe.copy()
    # Convertir la columna de fechas en un índice de tipo DatetimeIndex
    df['Date'] = pd.to_datetime(df['Date'])
    df.set_index('Date', inplace=True)
    #dataframe = dataframe.set_index('Date')
    df['Adj Close 2'] = df['Adj Close']
    
    # Crear un DataFrame vacío para almacenar los retornos mensuales
    monthly_returns_df = pd.DataFrame()

    # Agrupar los datos por símbolo de stock y calcular los retornos mensuales
    for symbol, group in df.groupby('Ticker'):
        monthly_data = group.resample('M').agg({'Adj Close': 'first', 'Adj Close 2': 'last'})
        monthly_data['Monthly_Return'] = ((monthly_data['Adj Close 2'] / monthly_data['Adj Close']) - 1) * 100
        monthly_data['Ticker'] = symbol  # Agregar una columna para el símbolo del stock
        monthly_returns_df = pd.concat([monthly_returns_df, monthly_data])

    monthly_return_by_stock = monthly_returns_df.groupby('Ticker').agg('mean')
    monthly_return_by_stock.drop(columns=['Adj Close', 'Adj Close 2'], inplace=True)

    return monthly_return_by_stock

```

#### Calcular la volatilidad ATR por stock

```python
# Calcular la volatilidad de cada stock con la formula de average true range
''' Average True Range (ATR)
		TR = max(High - Low, abs(High - Previous Close), abs(Low - Previous Close))
			* Luego sacar el promedio -> ese es el ATR'''
import pandas as pd

def calculate_volatility_ATR(dataframe):
	df = dataframe.copy()

    # Calculate True Range (TR) for each row in the DataFrame
	def calculate_tr(df):
		df['HL'] = df['High'] - df['Low']
		df['HC'] = abs(df['High'] - df['Close'].shift(1))
		df['LC'] = abs(df['Low'] - df['Close'].shift(1))
		df['TR'] = df[['HL', 'HC', 'LC']].max(axis=1)
    
    	# Drop the intermediate columns
		df.drop(['HL', 'HC', 'LC'], axis=1, inplace=True)

		# Calculate the Average True Range (ATR) using a rolling window
    	# period = 
    	# df['ATR'] = df['TR'].rolling(window=period, min_periods=1).mean() -- Lo dejo acá por si queremos calcularlo en distintos periodos

		return df
	
	#Calculate the average by stock. The period is the whole dataset. 
	tr_df = calculate_tr(df)
	atr_df = pd.DataFrame(tr_df.groupby('Ticker')['TR'].agg('mean'))
	atr_df.rename(columns={'TR':'ATR'}, inplace=True)

	return atr_df
```

```python
volatilidad_total = calculate_volatility_ATR(df_filtered)
volatilidad_total.head()
```
